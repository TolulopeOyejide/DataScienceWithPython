#### WHO Time Series

```python
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
who_time_series = pd.read_csv('WHO_time_series.csv')
who_time_series
# A series of data points listed in time is called 'Time Series'
# Time series are usually visualized with 'Line Graphs'
```

```python
who_time_series['Date_reported'] = pd.to_datetime(who_time_series['Date_reported'])
who_time_series
```

```python
who_time_series.info()
```

```python
#Visualizing Italy Data
italy = who_time_series[who_time_series['Country'] == 'Italy']
plt.plot(italy['Date_reported'], italy['Cumulative_cases'])
plt.title('Italy:Cummulative Reported Cases')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
```

A quantity that increases very quickly in beginning and then slows down more and more over time has LOGISTIC GROWTH

```python
#Visualizing Indian Data
India = who_time_series[who_time_series['Country'] == 'India']
plt.plot(India['Date_reported'], India['Cumulative_cases'])
plt.title('India:Cummulative Reported Cases')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
```

A quantity that increases very slowly in beginning and then grows faster and faster over time has EXPONENTIAL GROWTH

```python
#Creating a function that visualizes data for each country
def plot_cumulative_cases(country_name):
  country = who_time_series[who_time_series['Country'] == country_name]
  plt.plot(country['Date_reported'], country['Cumulative_cases'])
  plt.title('{}:Cummulative Reported Cases')
  plt.xlabel('Date')
  plt.ylabel('Number of Cases')
  plt.show()
```

```python
#Visualizing Poland Data
Poland = who_time_series[who_time_series['Country'] == 'Poland']
plt.plot(Poland['Date_reported'], Poland['Cumulative_cases'])
plt.title('Poland :Cummulative Reported Cases')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
```

A Quantity that increases constantly overtime is called is LINEAR GROWTH

```python
#Visualizing Belarus Data
Belarus = who_time_series[who_time_series['Country'] == 'Belarus']
plt.plot(Belarus['Date_reported'], Belarus['Cumulative_cases'])
plt.title('Belarus :Cummulative Reported Cases')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
```

```python
#Visualizing UK and France Data
France = who_time_series[who_time_series['Country'] == 'France']
UK = who_time_series[who_time_series['Country'] == 'The United Kingdom']
```

```python
plt.plot(France['Date_reported'], France['Cumulative_cases'], label = 'France')
plt.plot(UK['Date_reported'], UK['Cumulative_cases'], label = 'The UK')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.show()
```
